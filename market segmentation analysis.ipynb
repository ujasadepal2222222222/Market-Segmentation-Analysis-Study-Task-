```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x = pd.read_csv('mcdonalds.csv')
```

```python
l = ["yummy","convenient","spicy",'fattening','greasy','fast','cheap','tasty','expensive','healthy','disgusting']
```

```python
for i in l:
  x[i] = x[i].map({'Yes':1 ,'No':0})
```

```python
x
```

```python
x.corr()
```

```python
plt.pie(x['Gender'].value_counts(),  labels =  ['Female', 'Male'], shadow = True, autopct = '%.2f%%')
plt.title('Gender')
plt.legend()
plt.show()
```

```python
x['VisitFrequency'] = x['VisitFrequency'].replace(to_replace= 'Once a month', value = 3) 
x['VisitFrequency'] = x['VisitFrequency'].replace(to_replace= 'Every three months', value = 2) 
x['VisitFrequency'] = x['VisitFrequency'].replace(to_replace= 'Once a year', value = 1) 
x['VisitFrequency'] = x['VisitFrequency'].replace(to_replace= 'Once a week', value = 4) 
x['VisitFrequency'] = x['VisitFrequency'].replace(to_replace= 'Never', value = 0) 
x['VisitFrequency'] = x['VisitFrequency'].replace(to_replace= 'More than once a week', value = 5) 
```

```python
x.head()
```

```python
sns.countplot(x.Age)
plt.xlabel("Age")
plt.ylabel("Person Count")
plt.show()
```

```python
x['Gender'] = x['Gender'].map({'Male':1 ,'Female':0})
```

```python
y=[]
X= x.iloc[:, [13,14]].values

for i in range(1,15):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    y.append(kmeans.inertia_)
```
